### تکنیک های مورد استفاده قرار گرفته شده


در این پروژه از تکنیک ها و روش هایی استفاده شده تا هوش مصنوعی بازی کننده به بهترین عملکرد ممکن را انجام دهد و بهترین مسیر به سمت پیروز شدن را پیش بگیرد

در اینجا ابتدا هر مرحله ای که مورد بررسی توسط تابع minimax قرار میگیرد یک امتیاز تخصیص داده میشود که این امتیاز بسته به شرایط ممکن مقدایر و ضرایب متفاوتی خواهند داشت به عنوان مثال اگر در حرکتی موجب شود که ما به شرایط برد برسیم در حرکتی دیگر موجب شود که حریف از شرط بردن دور شود ما حالت اول برایمان اولویت بیشتری دارد چون به بردن تمایل بیشتری داریم تا به جلوگیری از باختن لذا ضریب بیشتری را به این حالت میدهیم 


تکنیک alpha , beta در این هوش مصنوعی استفاده شده است تا با حرص کردن برخی از شاخه های درخت سرعت تصمیم گیری هوش را افزایش بدهد.


و نیز در تکنیک دیگری که استفاده شده این است که هر مرحله دارای امتیازی است و علاوه بر امتیازی که توسط تابع min , max در مراحل پایینتر حساب میشود و به بالا بازگردانده میشود این امتیاز ها نیز حساب میشود تا از شرایط چند راهه شدن جلوگیری و شرایط چندراهه کردن حریف را بدست بیاورد.

در این بازی امکان اینک بتوان هوش مصنوعی را در حالت سخت یا آسان قرار داد وجود دارد که کاربر میتواند از بین انها یک را انتخاب کرده و با آن هوش بازی کند
و نیز در حالت AI vs AI نیز میتوان هر کدام از هوش مصنوعی هارا در یکی از حالت های سخت و آسان قرار داد و و بازی آنها را شروع کرد.
